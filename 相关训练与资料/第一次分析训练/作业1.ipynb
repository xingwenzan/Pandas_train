```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
divorce = pd.read_csv('divorce.csv')
marriage = pd.read_csv('marriage.csv')
income = pd.read_excel('income.xls')
```

```python
rows_name = ['area','2009','2010','2011','2012','2013','2014','2015','2016','2017']
years = ['2009','2010','2011','2012','2013','2014','2015','2016','2017']
year_num = [2009,2010,2011,2012,2013,2014,2015,2016,2017]
```

# 结论1

```python
x = divorce[years].columns
y = divorce[years].loc[0]
plt.plot(x, y)
plt.title('divorce of beijing') # plt 绘图不能有中文，真难受，网上教程都不好使，更难受了
plt.show()
```

> 结论1：北京离婚率总体成向上趋势，且在2016年达到最高

# 结论2

```python
plt.figure(12)

x = years

#plt.subplot(211)
y1 = divorce[years].loc[0]
y2 = marriage[years].loc[0]
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('divorce and marriage of beijing')

plt.show()
```

> 结论：北京结婚率始终高于离婚率

# 结论3

```python
y3 = income[year_num].loc[0]
plt.plot(x, y3)
plt.title('income of beijing')

plt.show()
```

>结论：北京收入几乎均匀增长

# 结论4

```python
print('编号地区对应表')
income[['area']]
```

```python
sizes = income[2009]
labels = income[2009].index

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, #阴影
        startangle=90)# 起始角
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
plt.bar(labels,sizes)
plt.title('income of area in 2009')
plt.show()
```

```python
income.sort_values(by = [2009],ascending = False)[['area',2009]]
```

> 结论：2009年上海人民收入最高,江西人民收入最低

# 结论5

```python
divorce.sort_values(by = ['2009'],ascending = False)[['area','2009']]
```

```python
marriage.sort_values(by = ['2009'],ascending = False)[['area','2009']]
```

# 结论6

```python
income['plus'] = income[2017] - income[2009]
```

```python
income.sort_values(by = ['plus'],ascending = False)[['area','plus']]
```
