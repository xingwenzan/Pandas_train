## 分组

### 1 - 分组统计｜均值

计算各区(`district`)的薪资(`salary`)均值

```python
df[['district','salary']].groupby('district').mean()
```

### 2 - 分组统计｜取消索引

重新按照上一题要求进行分组，但不使用 `district` 做为索引

```python
df[['district','salary']].groupby('district').mean().reset_index()
```

### 3 - 分组统计｜排序

计算并提取平均薪资最高的区

```python
df[['district','salary']].groupby('district').mean().sort_values('salary',ascending=False)
```

```python
df[['district','salary']].groupby('district').mean().reset_index().max()
```

```python
df[['district','salary']].groupby('district').mean().sort_values('salary',ascending=False).max()
```

```python
df[['district','salary']].groupby('district').mean().sort_values('salary',ascending=False).reset_index().max()
```

```python
df[['district','salary']].groupby('district').mean().sort_values('salary',ascending=False).head(1)
```

### 4 - 分组统计｜频率

计算不同行政区(`district`)，不同规模公司(`companySize`)出现的次数

```python
df[['district','companySize','salary']].groupby(by=['district','companySize']).count()
```

```python
pd.DataFrame(df.groupby("district")['companySize'].value_counts())

# df.groupby(['district','companySize']).size()
```

```python
df.groupby("district")['companySize']
```

```python
df.groupby("district")['companySize'].value_counts()
```

```python
df.groupby(['district','companySize']).size()
```

### 5 - 分组统计｜修改索引名

将上一题的索引名修改为
- district -> 行政区
- companySize -> 公司规模

```python
df[['district','companySize','salary']].groupby(by=['district','companySize']).count().rename_axis(["行政区", "公司规模"])
```

计算上一题，每个区出现的公司数量

```python
df.groupby('district')['companySize'].count()
```

### 8 - 分组查看｜指定

根据 createTime 列，计算每天不同 行政区 新增的岗位数量

```python
df[['createTime','district','salary']].groupby(by=['createTime','district']).count()
```

### 10 - 分组规则｜通过匿名函数2

计算各行政区的企业领域（industryField）包含电商的总数



```python
df[df['industryField']=='电商'].groupby('district')['salary'].count()
```

### 11 - 分组规则｜通过内置函数

通过 positionName 的长度进行分组，并计算不同长度岗位名称的薪资均值

```python
df.set_index("positionName")
```

```python
df.set_index("positionName").groupby(len)
```

```python
df.set_index("positionName").groupby(len)['salary'].mean()
```

### 12 - 分组规则｜通过字典

将 score 和 matchScore 的和记为总分，与 salary 列同时进行分组，并查看结果

```python
df.groupby({'salary':'薪资','score':'总分','matchScore':'总分'}, axis=1).sum()
```

### 13 - 分组规则｜通过多列

计算不同 工作年限（`workYear`）和 学历（`education`）之间的薪资均值

```python
df[['workYear','education','salary']].groupby(by=['workYear','education']).mean()
```

### 14 - 分组转换｜ transform

在原数据框 df 新增一列，数值为该区的平均薪资水平

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transform.html

```python
df['该区平均工资'] = df[['district','salary']].groupby(by='district').transform('mean')
```

```python
df[['district','salary']].groupby(by='district').transform('mean')
```

### 15 - 分组过滤｜filter


提取平均工资小于 30000 的行政区的全部数据

正则表达式 https://www.runoob.com/regexp/regexp-tutorial.html

```python
df.groupby('district').filter(lambda x: x['salary'].mean() < 30000)
```

```python
df[['district','salary']].groupby(by='district').transform('mean') < 30000
```

```python
df[(df[['district','salary']].groupby(by='district').transform('mean') < 30000)['salary']]
```

## 聚合

### 17 - 聚合统计

分组计算不同行政区，薪水的最小值、最大值和平均值

```python
import numpy as np
```

```python
df[['district','salary']].groupby('district').agg(['min','max','mean'])
```

```python
df[['district','salary']].groupby('district').agg([min, max, np.mean])
```

```python
df.groupby('district')['salary'].agg([min, max, np.mean])
```

### 18 - 聚合统计｜修改列名

将上一题的列名（包括索引名）修改为中文

```python
df.groupby('district').agg(最低工资=('salary', 'min'), 最高工资=('salary', 'max'), 平均工资=('salary', 'mean')).rename_axis(["行政区"])
```

### 19 - 聚合统计｜组合

对不同岗位(`positionName`)进行分组，并统计其薪水(`salary`)中位数和得分(`score`)均值

```python
df.groupby('positionName').agg(薪水=('salary','median'),得分=('score','mean'))
```

```python
df.groupby('positionName').agg({'salary': np.median, 'score': np.mean})
```

### 20 -聚合统计｜多层

对不同行政区进行分组，并统计薪水的均值、中位数、方差，以及得分的均值

```python
df.groupby('district').agg({'salary': [np.mean,np.median,np.std], 'score': np.mean})
```

```python
df.groupby('district').agg({'salary': ['mean','median','std'], 'score': 'mean'})
```

### 21 - 聚合统计｜自定义函数

在 18 题基础上，在聚合计算时新增一列计算最大值与平均值的差值

```python
def myfunc(x):

    return x.max()-x.mean()

df.groupby('district').agg(最低工资=('salary', 'min'), 最高工资=(
    'salary', 'max'), 平均工资=('salary', 'mean'), 最大值与均值差值=('salary', myfunc)).rename_axis(["行政区"])
```
